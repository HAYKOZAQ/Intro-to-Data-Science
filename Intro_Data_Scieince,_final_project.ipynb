```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import ttest_1samp

# ------------------------------------------------------------
# Helper Functions
# ------------------------------------------------------------

def fetch_yahoo_finance_data(ticker, interval='1d', range_period='2y'):
    """
    Fetches historical stock data for a given ticker from Yahoo Finance.

    Parameters:
    - ticker (str): Stock symbol (e.g., 'AAPL' for Apple Inc.)
    - interval (str): Data interval (e.g., '1d' for daily data)
    - range_period (str): Time range (e.g., '2y' for two years)

    Returns:
    - df (pd.DataFrame): DataFrame containing historical stock data with 'Date' as index
    """
    # Construct the Yahoo Finance API URL
    url = f"https://query1.finance.yahoo.com/v8/finance/chart/{ticker}?interval={interval}&range={range_period}"
    headers = {"User-Agent": "Mozilla/5.0"}  # Set User-Agent to mimic a browser request
    response = requests.get(url, headers=headers)  # Send GET request to fetch data

    # Check if the request was successful
    if response.status_code != 200:
        raise Exception(f"Failed to fetch data for {ticker}. "
                        f"Status Code: {response.status_code}\nResponse: {response.text}")

    try:
        data = response.json()  # Parse JSON response
    except ValueError:
        print(f"Response not valid JSON for {ticker}:")
        print(response.text)
        raise

    # Extract data from JSON structure
    if 'chart' in data and 'result' in data['chart'] and data['chart']['result']:
        result = data['chart']['result'][0]
        timestamps = result.get('timestamp', [])  # Unix timestamps
        indicators = result.get('indicators', {}).get('quote', [{}])[0]  # OHLCV data

        # Check if data is present
        if len(timestamps) == 0 or 'close' not in indicators:
            raise Exception(f"No valid data found for {ticker} and {range_period}.")

        # Create DataFrame from the fetched data
        df = pd.DataFrame({
            'Date': pd.to_datetime(timestamps, unit='s'),  # Convert timestamps to datetime
            'Open': indicators.get('open', []),
            'High': indicators.get('high', []),
            'Low': indicators.get('low', []),
            'Close': indicators.get('close', []),
            'Volume': indicators.get('volume', [])
        })
        df.set_index('Date', inplace=True)  # Set 'Date' as the index
        return df
    else:
        raise Exception(f"Unexpected JSON structure or no results found for {ticker}.")

def clean_data(df, ticker, forward_fill=False):
    """
    Cleans the fetched DataFrame by performing a series of data cleaning steps:
    - Removing duplicates
    - Handling missing values
    - Removing irrelevant or erroneous data points
    - Handling outliers
    - Ensuring data consistency and integrity

    Parameters:
    - df (pd.DataFrame): DataFrame to be cleaned
    - ticker (str): Stock ticker symbol (for logging purposes)
    - forward_fill (bool): If True, forward fills missing values; otherwise, drops them

    Returns:
    - df (pd.DataFrame): Cleaned DataFrame
    """
    print(f"\n--- Data Cleaning for {ticker} ---")
    print("Initial DataFrame Info:")
    print(df.info())  # Display DataFrame info before cleaning
    print("\nInitial NaN Counts:")
    print(df.isna().sum())  # Display count of NaNs in each column

    # 1. Remove duplicate rows based on the index (Date)
    print("\nRemoving duplicate rows based on Date index...")
    initial_count = len(df)
    df = df[~df.index.duplicated(keep='first')]
    duplicates_removed = initial_count - len(df)
    print(f"Duplicates removed: {duplicates_removed}")

    # 2. Remove rows with any negative or zero prices or volume
    print("\nRemoving rows with negative or zero prices or volume...")
    initial_count = len(df)
    condition = (df['Open'] > 0) & (df['High'] > 0) & (df['Low'] > 0) & (df['Close'] > 0) & (df['Volume'] > 0)
    df = df[condition]
    invalid_rows_removed = initial_count - len(df)
    print(f"Invalid rows removed: {invalid_rows_removed}")

    # 3. Ensure numeric columns are correctly typed (float)
    print("\nEnsuring numeric columns are of float type...")
    numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric, set errors to NaN

    # 4. Handle missing values
    print("\nHandling missing values...")
    if forward_fill:
        print("Forward filling missing values...")
        df.fillna(method='ffill', inplace=True)
    else:
        print("Dropping rows with any NaN values...")
        df.dropna(inplace=True)

    print("\nDataFrame Info After Handling Missing Values:")
    print(df.info())  # Display DataFrame info after handling missing values
    print("\nNaN Counts After Handling Missing Values:")
    print(df.isna().sum())  # Display count of NaNs after handling missing values

    # 5. Remove outliers using the Interquartile Range (IQR) method for 'Close' prices
    print("\nRemoving outliers in 'Close' prices using IQR method...")
    Q1 = df['Close'].quantile(0.25)
    Q3 = df['Close'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    initial_count = len(df)
    df = df[(df['Close'] >= lower_bound) & (df['Close'] <= upper_bound)]
    outliers_removed = initial_count - len(df)
    print(f"Outliers removed: {outliers_removed}")

    # 6. Additional consistency checks (e.g., High >= max(Open, Close), Low <= min(Open, Close))
    print("\nPerforming consistency checks on High and Low prices...")
    initial_count = len(df)
    condition = (df['High'] >= df[['Open', 'Close']].max(axis=1)) & (df['Low'] <= df[['Open', 'Close']].min(axis=1))
    df = df[condition]
    inconsistent_rows_removed = initial_count - len(df)
    print(f"Inconsistent rows removed: {inconsistent_rows_removed}")

    # 7. Reset index if necessary (optional)
    # print("\nResetting index...")
    # df.reset_index(inplace=True)

    print("\nFinal DataFrame Info After All Cleaning Steps:")
    print(df.info())  # Final info after all cleaning steps
    print("\nFinal NaN Counts:")
    print(df.isna().sum())  # Final NaN counts
    print("--- End of Data Cleaning ---\n")

    return df

def calculate_RSI(series, period=14):
    """
    Calculates the Relative Strength Index (RSI) for a given price series.

    Parameters:
    - series (pd.Series): Series of prices (typically closing prices)
    - period (int): Number of periods to use for RSI calculation (default is 14)

    Returns:
    - rsi (pd.Series): RSI values
    """
    delta = series.diff()  # Calculate price differences
    gain = np.where(delta > 0, delta, 0)  # Gains (only positive differences)
    loss = np.where(delta < 0, -delta, 0)  # Losses (only negative differences)
    avg_gain = pd.Series(gain, index=series.index).rolling(window=period).mean()  # Average gain
    avg_loss = pd.Series(loss, index=series.index).rolling(window=period).mean()  # Average loss
    rs = avg_gain / avg_loss  # Relative Strength
    rsi = 100.0 - (100.0 / (1.0 + rs))  # RSI calculation
    return rsi

def calculate_bollinger_bands(series, period=20, num_std=2):
    """
    Calculates Bollinger Bands for a given price series.

    Parameters:
    - series (pd.Series): Series of prices (typically closing prices)
    - period (int): Number of periods to use for the moving average (default is 20)
    - num_std (int): Number of standard deviations for the bands (default is 2)

    Returns:
    - rolling_mean (pd.Series): Moving average
    - upper_band (pd.Series): Upper Bollinger Band
    - lower_band (pd.Series): Lower Bollinger Band
    """
    rolling_mean = series.rolling(window=period).mean()  # Moving average
    rolling_std = series.rolling(window=period).std()  # Rolling standard deviation
    upper_band = rolling_mean + num_std * rolling_std  # Upper Bollinger Band
    lower_band = rolling_mean - num_std * rolling_std  # Lower Bollinger Band
    return rolling_mean, upper_band, lower_band

def calculate_MACD(series, fastperiod=12, slowperiod=26, signalperiod=9):
    """
    Calculates the Moving Average Convergence Divergence (MACD) and its signal line.

    Parameters:
    - series (pd.Series): Series of prices (typically closing prices)
    - fastperiod (int): Fast EMA period (default is 12)
    - slowperiod (int): Slow EMA period (default is 26)
    - signalperiod (int): Signal line EMA period (default is 9)

    Returns:
    - macd (pd.Series): MACD line
    - signal (pd.Series): Signal line
    """
    exp1 = series.ewm(span=fastperiod, adjust=False).mean()  # Fast EMA
    exp2 = series.ewm(span=slowperiod, adjust=False).mean()  # Slow EMA
    macd = exp1 - exp2  # MACD line
    signal = macd.ewm(span=signalperiod, adjust=False).mean()  # Signal line
    return macd, signal

def plot_candlestick(df, ticker):
    """
    Plots a candlestick chart for the given stock data.

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'Open', 'High', 'Low', 'Close' prices
    - ticker (str): Stock ticker symbol for the title
    """
    if df.empty:
        print("No data available for candlestick plot.")
        return

    fig, ax = plt.subplots(figsize=(12,6))
    width = 0.6  # Width of the candlestick bars

    for idx, (date, row) in enumerate(df.iterrows()):
        color = 'green' if row['Close'] >= row['Open'] else 'red'  # Color based on price movement
        # Plot the high-low line
        ax.plot([date, date], [row['Low'], row['High']], color=color)
        # Plot the open price
        ax.plot([date - pd.Timedelta(width/2, 'D'), date + pd.Timedelta(width/2, 'D')],
                [row['Open'], row['Open']], color=color)
        # Plot the close price
        ax.plot([date - pd.Timedelta(width/2, 'D'), date + pd.Timedelta(width/2, 'D')],
                [row['Close'], row['Close']], color=color)

    ax.set_title(f"{ticker} Candlestick Chart (Daily)")
    ax.set_xlabel("Date")
    ax.set_ylabel("Price (USD)")

    # Annotate the highest closing price
    highest_close = df['Close'].max()
    highest_date = df['Close'].idxmax()
    ax.annotate(f'Highest Close: {highest_close:.2f}',
                xy=(highest_date, highest_close),
                xytext=(highest_date, highest_close*1.05),
                arrowprops=dict(facecolor='black', shrink=0.05),
                fontsize=10,
                bbox=dict(boxstyle="round", fc="w"))

    plt.show()

def backtest_moving_average_strategy(df):
    """
    Backtests a simple moving average crossover trading strategy.

    Strategy:
    - Buy when MA20 > MA50
    - Sell when MA20 <= MA50

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'Close', 'MA20', 'MA50' columns

    Returns:
    - final_return (float): Final cumulative return of the strategy
    """
    if df.empty or 'MA20' not in df.columns or 'MA50' not in df.columns:
        print("Not enough data to backtest moving average strategy.")
        return np.nan

    df = df.copy()  # Create a copy to avoid SettingWithCopyWarning

    # Generate trading signals: 1 for buy, 0 for hold/sell
    df['Signal'] = np.where(df['MA20'] > df['MA50'], 1, 0)

    # Calculate daily returns
    df['Returns'] = df['Close'].pct_change()

    # Strategy returns: multiply daily returns by the previous day's signal
    df['Strategy'] = df['Returns'] * df['Signal'].shift(1)

    # Calculate cumulative returns
    cumulative_returns = (1 + df['Strategy']).cumprod() - 1

    # Return the final cumulative return
    return cumulative_returns.iloc[-1] if len(cumulative_returns) > 0 else np.nan

def backtest_rsi_strategy(df, rsi_lower=30, rsi_upper=70):
    """
    Backtests an RSI-based trading strategy.

    Strategy:
    - Buy when RSI < rsi_lower (e.g., 30)
    - Sell when RSI > rsi_upper (e.g., 70)

    Parameters:
    - df (pd.DataFrame): DataFrame containing 'Close' and 'RSI' columns
    - rsi_lower (int): RSI threshold for oversold conditions (default is 30)
    - rsi_upper (int): RSI threshold for overbought conditions (default is 70)

    Returns:
    - final_return (float): Final cumulative return of the strategy
    """
    if df.empty or 'RSI' not in df.columns:
        print("Not enough data to backtest RSI strategy.")
        return np.nan

    df = df.copy()  # Create a copy to avoid SettingWithCopyWarning

    # Initialize signals: 1 for buy, -1 for sell, 0 for hold
    df['Signal'] = 0
    df.loc[df['RSI'] < rsi_lower, 'Signal'] = 1
    df.loc[df['RSI'] > rsi_upper, 'Signal'] = -1

    # Calculate daily returns
    df['Returns'] = df['Close'].pct_change()

    # Position is the previous day's signal
    df['Position'] = df['Signal'].shift(1).fillna(0)

    # Strategy returns
    df['Strategy'] = df['Returns'] * df['Position']

    # Calculate cumulative returns
    cumulative_returns = (1 + df['Strategy']).cumprod() - 1

    # Return the final cumulative return
    return cumulative_returns.iloc[-1] if len(cumulative_returns) > 0 else np.nan

def calculate_VaR(returns, confidence=0.95):
    """
    Calculates the Value-at-Risk (VaR) for a given return series.

    Parameters:
    - returns (pd.Series): Series of returns
    - confidence (float): Confidence level for VaR (default is 0.95)

    Returns:
    - var (float): VaR at the specified confidence level
    """
    non_na_returns = returns.dropna()  # Remove NaNs
    if len(non_na_returns) == 0:
        return np.nan

    sorted_returns = np.sort(non_na_returns)  # Sort returns
    index = int((1 - confidence) * len(sorted_returns))  # Determine VaR index
    return sorted_returns[index] if index < len(sorted_returns) else np.nan

def create_dashboard():
    """
    Placeholder function for creating interactive dashboards using frameworks like Streamlit or Dash.
    """
    pass

def schedule_daily_update():
    """
    Placeholder function for scheduling daily updates using tools like cron jobs or Airflow.
    """
    pass

# ------------------------------------------------------------
# Main Analysis Function
# ------------------------------------------------------------

def analyze_stock(ticker, benchmark_ticker='SPY', additional_ticker='MSFT'):
    """
    Performs comprehensive analysis and visualization for a given stock ticker.

    Parameters:
    - ticker (str): Stock symbol to analyze (e.g., 'AAPL', 'GOOGL', 'AMZN')
    - benchmark_ticker (str): Benchmark stock symbol (default is 'SPY')
    - additional_ticker (str): Additional stock symbol for correlation analysis (default is 'MSFT')
    """
    print(f"\n{'='*50}\nStarting analysis for {ticker}\n{'='*50}")

    # Fetch historical data for the main ticker, benchmark, and additional ticker
    try:
        stock_df = fetch_yahoo_finance_data(ticker, interval='1d', range_period='2y')
        benchmark_df = fetch_yahoo_finance_data(benchmark_ticker, interval='1d', range_period='2y')
        extra_df = fetch_yahoo_finance_data(additional_ticker, interval='1d', range_period='2y')
    except Exception as e:
        print(f"Error fetching data: {e}")
        return

    # Clean the fetched data with enhanced data cleaning
    stock_df = clean_data(stock_df, ticker, forward_fill=False)
    benchmark_df = clean_data(benchmark_df, benchmark_ticker, forward_fill=False)
    extra_df = clean_data(extra_df, additional_ticker, forward_fill=False)

    # Align all three DataFrames on common dates to ensure consistency
    common_dates = stock_df.index.intersection(benchmark_df.index).intersection(extra_df.index)
    stock_df = stock_df.loc[common_dates]
    benchmark_df = benchmark_df.loc[common_dates]
    extra_df = extra_df.loc[common_dates]

    # Check if there is overlapping data
    if stock_df.empty:
        print("No overlapping data among tickers.")
        return

    # Integrate Mock Fundamental Data (e.g., PE Ratio, EPS Growth)
    fundamental_data = pd.DataFrame({
        'Date': stock_df.index,
        'PE_Ratio': np.random.uniform(15, 30, size=len(stock_df)),        # Mock PE Ratios between 15 and 30
        'EPS_Growth': np.random.uniform(-0.1, 0.1, size=len(stock_df))   # Mock EPS Growth between -10% and +10%
    }).set_index('Date')
    stock_df = stock_df.join(fundamental_data)  # Join fundamental data to stock DataFrame

    # Integrate Mock Macroeconomic Indicator (e.g., Interest Rate)
    interest_rate = pd.DataFrame({
        'Date': stock_df.index,
        'Interest_Rate': np.random.uniform(0.5, 2.5, size=len(stock_df))  # Mock interest rates between 0.5% and 2.5%
    }).set_index('Date')
    stock_df = stock_df.join(interest_rate)  # Join macro data to stock DataFrame

    # Compute Technical Indicators for the main ticker
    stock_df['MA20'] = stock_df['Close'].rolling(window=20).mean()  # 20-day Moving Average
    stock_df['MA50'] = stock_df['Close'].rolling(window=50).mean()  # 50-day Moving Average
    stock_df['RSI'] = calculate_RSI(stock_df['Close'])              # Relative Strength Index
    stock_df['BB_MID'], stock_df['BB_UP'], stock_df['BB_LOW'] = calculate_bollinger_bands(stock_df['Close'])  # Bollinger Bands
    stock_df['MACD'], stock_df['MACD_Signal'] = calculate_MACD(stock_df['Close'])  # MACD and Signal Line

    # Add Mock Economic Indicator (e.g., general economic condition)
    np.random.seed(42)  # Set seed for reproducibility
    econ_indicator = np.random.normal(2.0, 0.1, size=len(stock_df))  # Mock economic indicator
    stock_df['Econ_Indicator'] = econ_indicator  # Add to DataFrame

    # Compute Daily Returns for the main ticker, benchmark, and additional ticker
    stock_df['Returns'] = stock_df['Close'].pct_change()          # Main ticker returns
    benchmark_df['Benchmark_Returns'] = benchmark_df['Close'].pct_change()  # Benchmark returns
    extra_df['Extra_Returns'] = extra_df['Close'].pct_change()    # Additional ticker returns

    # Drop NaN returns to ensure alignment
    stock_df.dropna(subset=['Returns'], inplace=True)
    benchmark_df.dropna(subset=['Benchmark_Returns'], inplace=True)
    extra_df.dropna(subset=['Extra_Returns'], inplace=True)

    # Realign DataFrames after calculating returns
    final_common = stock_df.index.intersection(benchmark_df.index).intersection(extra_df.index)
    stock_df = stock_df.loc[final_common]
    benchmark_df = benchmark_df.loc[final_common]
    extra_df = extra_df.loc[final_common]

    # Drop any remaining NaNs after realignment
    stock_df.dropna(inplace=True)
    benchmark_df.dropna(inplace=True)
    extra_df.dropna(inplace=True)

    # Calculate Change in Economic Indicator
    stock_df['Econ_Indicator_Change'] = stock_df['Econ_Indicator'].pct_change()
    stock_df.dropna(subset=['Econ_Indicator_Change'], inplace=True)

    # Portfolio-Level Analysis: Correlation Matrix
    all_returns = pd.DataFrame({
        ticker: stock_df['Returns'],
        benchmark_ticker: benchmark_df['Benchmark_Returns'],
        additional_ticker: extra_df['Extra_Returns']
    }).dropna()

    print(f"\nPortfolio-level correlation matrix for {ticker}:")
    corr_matrix = all_returns.corr()
    print(corr_matrix)

    # Plot Correlation Heatmap using Seaborn
    plt.figure(figsize=(6,4))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f"Correlation Heatmap of {ticker} Portfolio Returns")
    plt.show()

    # Backtesting RSI-Based Trading Strategy
    rsi_strategy_return = backtest_rsi_strategy(stock_df)
    print(f"RSI Strategy Final Cumulative Return for {ticker}: {rsi_strategy_return*100:.2f}%")

    # Backtesting Moving Average Crossover Trading Strategy
    ma_strategy_return = backtest_moving_average_strategy(stock_df)
    print(f"MA Crossover Strategy Final Cumulative Return for {ticker}: {ma_strategy_return*100:.2f}%")

    # Statistical Testing: T-test on Returns
    t_stat, p_val = ttest_1samp(stock_df['Returns'].dropna(), 0)
    print(f"\nT-test on returns for {ticker}: t-stat={t_stat:.4f}, p-value={p_val:.4f}")

    # Scenario and Sensitivity Analysis Example
    adjusted_interest_rate = stock_df['Interest_Rate'] + 0.5
    print(f"\nScenario Test for {ticker}: Increase interest rate by 0.5%. You could re-run your model with this adjusted rate.")

    # Modeling: Predict Next Day's Closing Price
    df_model = stock_df.copy()
    df_model['Target'] = df_model['Close'].shift(-1)  # Next-day close as target variable
    df_model.dropna(inplace=True)  # Drop the last row with NaN target

    # Define Features and Target
    X = df_model[['Close', 'MA20', 'MA50', 'RSI', 'BB_MID', 'BB_UP', 'BB_LOW', 'MACD', 'MACD_Signal',
                  'Econ_Indicator', 'PE_Ratio', 'EPS_Growth', 'Interest_Rate']]
    y = df_model['Target']

    # Split into Train and Test Sets (80-20 Split)
    split_index = int(len(df_model)*0.8)
    X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
    y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

    # Check if Train and Test Sets are Valid
    if len(X_train) == 0 or len(X_test) == 0:
        print("Train or test set is empty. Increase the data range.")
        return

    # Fit Linear Regression Model
    print(f"\nFitting Linear Regression for {ticker}...")
    lr_model = LinearRegression()
    lr_model.fit(X_train, y_train)  # Train the model
    y_pred_lr = lr_model.predict(X_test)  # Predict on test set
    mae_lr = np.mean(np.abs(y_test - y_pred_lr))  # Calculate Mean Absolute Error
    print(f"Linear Regression MAE for {ticker}: {mae_lr:.4f} USD")

    # Fit Random Forest Regressor Model
    print(f"Fitting Random Forest Regressor for {ticker}...")
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)  # Train the model
    y_pred_rf = rf_model.predict(X_test)  # Predict on test set
    mae_rf = np.mean(np.abs(y_test - y_pred_rf))  # Calculate Mean Absolute Error
    print(f"Random Forest MAE for {ticker}: {mae_rf:.4f} USD")

    # Compare Actual vs Predicted Prices
    comparison_df = pd.DataFrame({'Actual': y_test, 'LR_Pred': y_pred_lr, 'RF_Pred': y_pred_rf}, index=y_test.index)
    print(f"\nSample predictions (Actual vs. Model Predictions) for {ticker}:")
    print(comparison_df.head())

    # Plot Actual vs Predicted Prices
    plt.figure(figsize=(14,7))
    plt.plot(comparison_df.index, comparison_df['Actual'], label='Actual', color='black')
    plt.plot(comparison_df.index, comparison_df['LR_Pred'], label='Linear Regression Prediction', color='blue', alpha=0.7)
    plt.plot(comparison_df.index, comparison_df['RF_Pred'], label='Random Forest Prediction', color='green', alpha=0.7)
    plt.title(f"Actual vs Predicted Close Prices for {ticker}")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.show()

    # Plot Candlestick Chart for the main ticker
    print(f"\nPlotting candlestick chart for {ticker}...")
    plot_candlestick(stock_df, ticker)

    # Plot Histogram of Daily Returns
    plt.figure(figsize=(8,5))
    plt.hist(stock_df['Returns'].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.title(f"Distribution of {ticker} Daily Returns")
    plt.xlabel("Daily Return")
    plt.ylabel("Frequency")
    plt.show()

    # Scatter Plot: Main Ticker Returns vs Benchmark Returns
    print(f"\nPlotting returns vs. benchmark returns scatter plot for {ticker}...")
    # Create a DataFrame for scatter plot data
    scatter_data = pd.DataFrame({
        'Benchmark_Returns': benchmark_df['Benchmark_Returns'],
        'Stock_Returns': stock_df['Returns']
    }).dropna()

    # Verify that both Series have the same length after dropping NaNs
    if scatter_data['Benchmark_Returns'].size != scatter_data['Stock_Returns'].size:
        print("x and y must be the same size after alignment.")
    else:
        # Calculate correlation for annotation
        corr_value = scatter_data['Benchmark_Returns'].corr(scatter_data['Stock_Returns'])

        # Create Scatter Plot
        fig, ax = plt.subplots(figsize=(8,6))
        ax.scatter(scatter_data['Benchmark_Returns'], scatter_data['Stock_Returns'], alpha=0.5)
        ax.set_xlabel(f"{benchmark_ticker} Daily Returns")
        ax.set_ylabel(f"{ticker} Daily Returns")
        ax.set_title(f"Scatter Plot of {ticker} Daily Returns vs. {benchmark_ticker}")
        # Annotate correlation coefficient on the plot
        ax.text(0.05, 0.95, f"Correlation: {corr_value:.2f}", transform=ax.transAxes, fontsize=12,
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))
        plt.show()

    # Additional Dedicated Plots

    # 1. Relative Strength Index (RSI) Chart
    plt.figure(figsize=(14,5))
    plt.plot(stock_df.index, stock_df['RSI'], label='RSI', color='purple')
    plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
    plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
    plt.title(f"{ticker} Relative Strength Index (RSI)")
    plt.xlabel("Date")
    plt.ylabel("RSI")
    plt.legend()
    plt.show()

    # 2. Moving Average Convergence Divergence (MACD) Chart
    plt.figure(figsize=(14,5))
    plt.plot(stock_df.index, stock_df['MACD'], label='MACD', color='blue')
    plt.plot(stock_df.index, stock_df['MACD_Signal'], label='Signal Line', color='red')
    plt.title(f"{ticker} Moving Average Convergence Divergence (MACD)")
    plt.xlabel("Date")
    plt.ylabel("MACD")
    plt.legend()
    plt.show()

    # 3. Price with Moving Averages and Bollinger Bands Chart
    plt.figure(figsize=(14,7))
    plt.plot(stock_df.index, stock_df['Close'], label='Close Price', color='black')
    plt.plot(stock_df.index, stock_df['MA20'], label='MA20', color='blue', alpha=0.7)
    plt.plot(stock_df.index, stock_df['MA50'], label='MA50', color='red', alpha=0.7)
    plt.fill_between(stock_df.index, stock_df['BB_UP'], stock_df['BB_LOW'], color='grey', alpha=0.3, label='Bollinger Bands')
    plt.title(f"{ticker} Price with MA20, MA50, and Bollinger Bands")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.show()

    # 4. Simple Line Plot of Closing Prices (Stock Price Trend)
    plt.figure(figsize=(14,7))
    plt.plot(stock_df.index, stock_df['Close'], label='Close Price', color='blue')
    plt.title(f"{ticker} Closing Price Over Time")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.show()

    # 5. Line Plot with Moving Averages to Highlight Trends
    plt.figure(figsize=(14,7))
    plt.plot(stock_df.index, stock_df['Close'], label='Close Price', color='black', alpha=0.5)
    plt.plot(stock_df.index, stock_df['MA20'], label='20-Day MA', color='blue')
    plt.plot(stock_df.index, stock_df['MA50'], label='50-Day MA', color='red')
    plt.title(f"{ticker} Closing Price with 20-Day and 50-Day Moving Averages")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.show()

    # 6. Trend Lines Using Linear Regression
    # Fit a linear regression model to identify the overall trend
    plt.figure(figsize=(14,7))
    plt.plot(stock_df.index, stock_df['Close'], label='Close Price', color='black', alpha=0.5)

    # Convert dates to ordinal for regression
    X_dates = stock_df.index.map(datetime.toordinal).values.reshape(-1,1)
    y_prices = stock_df['Close'].values

    # Fit linear regression
    lr_trend = LinearRegression()
    lr_trend.fit(X_dates, y_prices)
    trend_line = lr_trend.predict(X_dates)

    # Plot trend line
    plt.plot(stock_df.index, trend_line, label='Trend Line (Linear Regression)', color='orange', linewidth=2)
    plt.title(f"{ticker} Closing Price with Trend Line")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.show()

    # Risk Analysis: Value-at-Risk (VaR)
    var_95 = calculate_VaR(stock_df['Returns'], confidence=0.95)
    if pd.notnull(var_95):
        print(f"\n95% VaR for {ticker}: {var_95*100:.2f}% (daily)")
    else:
        print(f"\nNot enough data to calculate VaR for {ticker}.")

    print(f"\n{'='*50}\nCompleted analysis for {ticker}\n{'='*50}\n")

# ------------------------------------------------------------
# Main Function to Analyze Multiple Tickers
# ------------------------------------------------------------

def main():
    """
    Main function to perform comprehensive stock analysis for multiple tickers.
    """
    # Define the list of tickers to analyze
    tickers = ['AAPL', 'GOOGL', 'AMZN']  # Apple, Google, Amazon

    # Define benchmark and additional tickers
    benchmark_ticker = 'SPY'  # S&P 500 ETF as benchmark
    additional_ticker = 'MSFT'  # Microsoft Corporation as additional ticker

    # Iterate through each ticker and perform analysis
    for ticker in tickers:
        analyze_stock(ticker, benchmark_ticker=benchmark_ticker, additional_ticker=additional_ticker)

    # Placeholder calls for dashboards and scheduling (to be implemented as needed)
    # create_dashboard()
    # schedule_daily_update()

# ------------------------------------------------------------
# Entry Point
# ------------------------------------------------------------
if __name__ == "__main__":
    main()
```
